```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#linspace 100 valores entre 0 y 10
x = np.linspace(0,10,100)
y = np.sin(x)
z= np.cos(x)
```

```python
#ola seno
plt.plot(x,y)
plt.show()
```

```python
#ola coseno
plt.plot(x,z)
plt.show()
```

```python
#agregar TITULO
plt.plot(x,y)
plt.xlabel('angle')
plt.ylabel('sign value')
plt.title('ola seno')
plt.show()
```

```python
x=np.linspace(-10,10,20)
x
y=x**2
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y, 'r+')
plt.show()
```

```python
plt.plot(x,y, 'rx')
plt.show()
```

```python
q= np.linspace(-5,5,50)
q
#g- es color verde
plt.plot(q, np.sin(q), color='g')
plt.plot(q, np.cos(q), 'r--')
```

BAR PLOT

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
languages=['English', 'French', 'Spanish', 'Latin', 'German']
people=[100,50,150,40,70]
ax.bar(languages, people)
plt.xlabel('Languages')
plt.ylabel('number of people')
plt.show()
```

```python
fig1= plt.figure()
ax = fig1.add_axes([0,0,1,1])
languages=['English', 'French', 'Spanish', 'Latin', 'German']
people=[100,50,150,40,70]
ax.pie(people, labels = languages, autopct='%1.1f%%')
plt.show()
```

```python
x=np.linspace(0,10,30)
y=np.sin(x)
z=np.cos(x)
fig2= plt.figure()
ax = fig2.add_axes([0,0,1,1])
ax.scatter(x,y, color='g')
ax.scatter(x,z, color='b')
plt.show()
```

3D

```python
fig3=plt.figure()
ax= plt.axes(projection='3d')
z=20 * np.random.random(100)
x=np.sin(z)
y=np.cos(z)
ax.scatter(x,y,z,  cmap='blue')
plt.show()
```
